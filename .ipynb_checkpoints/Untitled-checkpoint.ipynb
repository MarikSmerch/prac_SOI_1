```python
import numpy as np
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns

# Преобразование
df['log_norm'] = np.log(df['normalized'])

# Гистограмма и плотность логарифмированных данных
plt.figure(figsize=(10,6))
sns.histplot(df['log_norm'], bins=12, kde=True, color='lightgreen', edgecolor='black')
plt.title('Распределение логарифмированных данных ln(X): число больничных организаций на 100 000 населения')
plt.xlabel('ln(числа организаций на 100 000 населения)')
plt.ylabel('Плотность вероятности')
plt.grid(True, alpha=0.3)
plt.show()

# Проверка нормальности после логарифмирования
stat, p_value = shapiro(df['log_norm'])
print(f"Статистика Шапиро–Уилка: {stat:.4f}, p-value: {p_value:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка нормализованных данных
df = pd.read_csv('bolnich_2023_normalized.csv')

# Построение boxplot (ящика с усами)
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['normalized'], color='lightblue', width=0.4, fliersize=5, boxprops=dict(alpha=0.7))

plt.title('Boxplot распределения числа больничных организаций на 100 000 населения, 2023 г.', fontsize=14)
plt.ylabel('Число больничных организаций', fontsize=12)
plt.grid(True, axis='y', alpha=0.3)

# Добавляем подписи к основным элементам
plt.text(0.05, df['normalized'].median() + 0.05, f'Медиана = {df["normalized"].median():.2f}', fontsize=10)
plt.text(0.05, df['normalized'].quantile(0.75) + 0.1, f'Q3 = {df["normalized"].quantile(0.75):.2f}', fontsize=10)
plt.text(0.05, df['normalized'].quantile(0.25) - 0.4, f'Q1 = {df["normalized"].quantile(0.25):.2f}', fontsize=10)

plt.savefig('boxplot_bolnich_2023.png', dpi=300, bbox_inches='tight')
plt.show()
```
