```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Загрузка файлов
df22 = pd.read_csv('bolnich_2022_normalized.csv')
df23 = pd.read_csv('bolnich_2023_normalized.csv')

# Объединение
merged = pd.merge(df22[['region','normalized']],
                  df23[['region','normalized']],
                  on='region', suffixes=('_2022','_2023'))

# Корреляция
r, _ = pearsonr(merged['normalized_2022'], merged['normalized_2023'])
print(f'Коэффициент корреляции Пирсона: r={r:.3f}')

# Диаграмма рассеяния
plt.figure(figsize=(8,6))
plt.scatter(merged['normalized_2022'], merged['normalized_2023'],
            color='skyblue', alpha=0.7, edgecolor='black')
plt.plot([0, merged[['normalized_2022','normalized_2023']].max().max()],
         [0, merged[['normalized_2022','normalized_2023']].max().max()],
         color='red', linestyle='--', label='линия равенства y = x')
plt.title('Сравнение числа больничных организаций на 100 000 населения: 2022 и 2023 гг.')
plt.xlabel('2022 год')
plt.ylabel('2023 год')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('scatter_2022_2023.png', dpi=300)
plt.show()
```

```python
# Расчёт изменений
merged['delta'] = merged['normalized_2023'] - merged['normalized_2022']
merged['delta_percent'] = (merged['delta'] / merged['normalized_2022']) * 100

# Средние изменения
mean_delta = merged['delta'].mean()
mean_delta_percent = merged['delta_percent'].mean()
print(f"Среднее изменение Δ = {mean_delta:.3f}")
print(f"Среднее относительное изменение = {mean_delta_percent:.2f}%")

# Гистограмма
plt.figure(figsize=(9,6))
plt.hist(merged['delta'], bins=15, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(0, color='red', linestyle='--', label='нулевое изменение')
plt.title('Гистограмма изменений числа больничных организаций на 100 000 населения (2023–2022)')
plt.xlabel('Изменение Δ (ед.)')
plt.ylabel('Частота (число регионов)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('hist_delta_2022_2023.png', dpi=300)
plt.show()
```
