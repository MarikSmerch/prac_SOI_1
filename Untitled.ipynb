```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка нормализованных данных
df = pd.read_csv('bolnich_2023_normalized.csv')

# Проверка максимального значения для диапазона
max_normalized = df['normalized'].max()
print(f"Максимальное нормализованное значение: {max_normalized}")

# Гистограмма
plt.figure(figsize=(10, 6))
plt.hist(df['normalized'], bins=10, range=(0, max(max_normalized + 5, 25)), edgecolor='black', alpha=0.6, color='skyblue', label='Частота субъектов')
plt.title('Распределение числа больничных организаций на 100 000 населения по субъектам РФ, 2023 год', fontsize=14)
plt.xlabel('Число больничных организаций на 100 000 населения, ед.', fontsize=12)
plt.ylabel('Частота (число субъектов)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('histogram_normalized_2023.png', dpi=300, bbox_inches='tight')
plt.show()

# Топ-10 регионов
top_10 = df.nlargest(10, 'normalized')
plt.figure(figsize=(12, 6))
plt.bar(top_10['region'], top_10['normalized'], color='lightgreen', label='Обеспеченность')
plt.title('Топ-10 регионов по обеспеченности больницами на 100 000 населения, 2023 год', fontsize=14)
plt.xlabel('Регионы', fontsize=12)
plt.ylabel('Число больничных организаций на 100 000 населения, ед.', fontsize=12)
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.savefig('top_10_2023.png', dpi=300, bbox_inches='tight')
plt.show()

# Бот-10 регионов
bot_10 = df.nsmallest(10, 'normalized')
plt.figure(figsize=(12, 6))
plt.bar(bot_10['region'], bot_10['normalized'], color='lightcoral', label='Обеспеченность')
plt.title('Бот-10 регионов по обеспеченности больницами на 100 000 населения, 2023 год', fontsize=14)
plt.xlabel('Регионы', fontsize=12)
plt.ylabel('Число больничных организаций на 100 000 населения, ед.', fontsize=12)
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3)
plt.savefig('bot_10_2023.png', dpi=300, bbox_inches='tight')
plt.show()
```
